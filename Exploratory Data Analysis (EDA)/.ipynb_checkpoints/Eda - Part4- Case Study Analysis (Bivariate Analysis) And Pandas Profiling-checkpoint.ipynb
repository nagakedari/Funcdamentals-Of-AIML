# Bivariate Analysis


Through bivariate analysis we try to analyze two variables simultaneously. As opposed to univariate analysis where we check the characteristics of a single variable, in bivariate analysis we try to determine if there is any relationship between two variables.

 

There are essentially 3 major scenarios that we will come accross when we perform bivariate analysis
<ol>
<li>Both variables of interest are qualitative </li>
<li>One variable is qualitative and the other is quantitative </li>
<li>Both variables are quantitative</li>
    </ol>
 

For the purpose of this exercise, we will explore few most popular techniques to perform bivariate analysis.

The following plots are not limited to the headings they are under. They are the options we have if we face a certain scenario.

# Numerical vs. Numerical
1. Scatterplot
2. Line plot
3. Heatmap for correlation
4. Joint plot

# Categorical vs. Numerical
1. Bar chart
2. Voilin plot
3. Categorical box plot
4. Swarm plot

# Two Categorical Variables
1. Bar chart
2. Grouped bar chart
3. Point plot

# Numerical Vs Numerical

```python
figure = plt.figure(figsize=(15,5))
sns.scatterplot(x='population', y='suicides_no', data=data, size='suicides_no')
```

```python
figure = plt.figure(figsize=(50,15))
sns.regplot(x='population', y='suicides_no', data=data)
```

<b>Note:</b> the regression x on y is not same as y on x

```python
figure = plt.figure(figsize=(50,15))
sns.lmplot(x='population', y='suicides_no', data=data)
```

```python
sns.lineplot(x='population', y='suicides_no', data=data.head())
```

```python
plt.figure(figsize=(15,5))
sns.scatterplot(x='gdp_per_capita ($)', y='suicides/100k pop', data=data)
```

<li>It looks like higher suicide rates are bit relevant to countries with lower GDP per capita

```python
plt.figure(figsize=(10,5))
sns.heatmap(data.corr(), annot=True, linewidths=.5, fmt= '.1f', center = 1 )  # heatmap
plt.show()
```

<li>Instead of doing a simple df.corr(), we can use the heatmap when there are large number of variables
<li>The color helps with picking out the most correlated easily
<li>In the above case, we don't have a lot of variables the plot is just for demonstration
<li>The darker the color higher the correlation
<li>None of the attributes seem have correlation of real significance
<li>Some obvious correlations are that in a larger population, it is very likely that the number of suicides will be more
<li>Human Development Index - gdp per capita is the only pair with the most correlation

# Categorical Vs Numerical

```python
plt.figure(figsize=(15,5))
sns.barplot(x='sex', y='suicides_no', data=data, hue='age')
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='sex', y='suicides_no', data=data, hue='generation')
```

```python
plt.figure(figsize=(15,5))
plt.title('Count of suicides for 31 years')
sns.barplot(x='suicides_no', y='country', data=data.groupby([data['country'], data['sex']]).sum().reset_index().sort_values(by='suicides_no', ascending=False).head(10), hue='sex')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.pointplot(x='generation', y='suicides_no', hue='sex', data=data)
plt.show()
```

`generally we will go for point plots if we wanted to see y value over a period of time, the above is a point plot of no. of suicides over generations`

```python
plt.figure(figsize=(15,5))
sns.violinplot(x='generation', y='population', data=data)
```

` The violing plot doesn't give any additional information, but since it display the symetric, we can easily see the variation from left to right as human eye is comfirtable looking at left to right`

<li>The plot is similar to a box plot but here, we get a density function</li>
<li>Distribution of population accross every eneration is highly skewed</li>
<li>Potential of lot of outliers</li>

```python
plt.figure(figsize=(15,5))
sns.violinplot(x='generation', y='suicides_no', data=data, hue='country')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x='generation', y='population', data=data)
```

# Checking trends with Temporal Data

`Temporal data is simply data that represents a state in time, such as the land-use patterns of Hong Kong in 1990, or total rainfall in Honolulu on July 1, 2009. Temporal data is collected to analyze weather patterns and other environmental variables, monitor traffic conditions, study demographic trends, and so on. This data comes from many sources ranging from manual data entry to data collected using observational sensors or generated from simulation models.`

```python
plt.figure(figsize=(15,5))
# data.groupby(by='year').sum().reset_index()
sns.lineplot(x='year', y='suicides_no', data=data.groupby(by='year').sum().reset_index())
```

`While comparing temporal data we need to make sure the time is symmetric (which means if we have lets say 10 years of data and if the last year data is not for full year its not symmetric on time)`

```python
plt.figure(figsize=(15,5))
sns.lineplot(x='year', y='population', data=data.groupby(by='year').sum().reset_index())
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(x='year', y='suicides/100k pop', data=data.groupby(by='year').sum().reset_index())
```
